```python
import sys 
sys.path.append("..")

import numpy as np
import src.tools.diffusion.black_scholes as bs
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
```

### Check get_path and get_path_correlated function

```python
asset = bs.BlackScholes(36, 0.06, 0.2)
rho = 0.8
corr_matrix = [[1, rho], [rho, 1]]
n = 1000
dt = 1/n
N = 2
```

```python
simulations = asset.get_path(n,dt,N)
plt.plot(simulations[0,:])
plt.plot(simulations[1,:])
plt.title("independent paths")
```

```python
simulations = asset.get_path_correlated(n,dt,N,corr_matrix)
plt.plot(simulations[0,:])
plt.plot(simulations[1,:])
plt.title(f"correlated paths with $\\rho$ ={rho}")
```

```python
rho = -0.9
corr_matrix = [[1, rho], [rho, 1]]
simulations = asset.get_path_correlated(n,dt,N,corr_matrix)
plt.plot(simulations[0,:])
plt.plot(simulations[1,:])
plt.title(f"correlated paths with $\\rho$ ={rho}")
```
